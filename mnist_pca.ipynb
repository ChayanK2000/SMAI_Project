```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
import sys
import math
import keras
from keras.datasets import mnist
```

# Loading the Dataset

```python
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
print(train_X.shape)

new_X = [np.reshape(train_X[i],784) for i in range(len(train_X))]
print(len(new_X[0]))
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
print(train_X.shape)

new_X_train = sc.fit_transform(new_X)
print(len(new_X_train[0]))
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


pca = PCA(n_components = 2)
 
Y = pca.fit_transform(new_X_train)
# X_test = pca.transform(X_test)
 
# explained_variance = pca.explained_variance_ratio_

# print("initial data:\n ",train_X[0])
# print("after scaling and pca: \n",type(new_X_train[0]))
# print("label: \n",train_y[0])
# print(len(new_X_train))
# digits_final = TSNE(perplexity=10).fit_transform(new_X_train[:2000])
# print(digits_final)
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame()
x = [i[0] for i in Y[:200]]
y = [i[1] for i in Y[:200]]
df["feat-1"] = x
df["feat-2"] = y
cat=[]
for i in train_y[:200]:
  cat.append(i)
# df["y"] = cat

sns.scatterplot(x="feat-1", y="feat-2", hue=cat,
                palette=sns.color_palette("hls", 10),
                data=df).set(title="PCA clustering on MNIST")
```
