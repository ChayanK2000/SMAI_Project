```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_digits
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
```

```python
# DATASET PREPARATION

digits = load_digits()
print([i for i in digits.target])

new_X = digits.data
train_y =digits.target
```

```python
# SCALING THE DATASET AND REDUCING DIMENSIONS
sc = StandardScaler()

new_X_train = sc.fit_transform(new_X)
pca = PCA(n_components = 30)
new_X_train = pca.fit_transform(new_X_train)

print(new_X_train.shape)
```

```python
# TRANSFORMING THE DATASET USING Locally Linear Embedding DIMENSIONALITY REDUCTION
model = LocallyLinearEmbedding(n_components=2)
X_lle = model.fit_transform(new_X_train)
print(X_lle.shape)
```

```python
# VISUALIZING THE DATASET RETURNED BY LLE TECHINIQUE
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=train_y, s=8)
plt.title('Locally Linear Embedding Visualization')
plt.rcParams['figure.figsize'] = [20, 20]

plt.show()
```
