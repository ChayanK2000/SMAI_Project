```python
# IMPORTING THE REQUIRED LIBRARIES
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# DATASET PREPARATION
digits = load_digits()
print([i for i in digits.target])

new_X = digits.data
train_y =digits.target
```

```python
# SCALING THE DATASET
sc = StandardScaler()
new_X_train = sc.fit_transform(new_X)
```

```python
# TRANSFORMING THE DATASET USING PCA DIMENSIONALITY REDUCTION
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(new_X_train)

print(X_pca.shape)
```

```python
# VISUALIZING THE DATASET RETURNED BY PCA TECHINIQUE
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=train_y, s=8)
plt.title('PCA Visualization')
plt.rcParams['figure.figsize'] = [10, 10]

plt.show()
```
