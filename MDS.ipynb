```python
# IMPORTING THE REQUIRED LIBRARIES
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

```python
# DATASET PREPARATION
digits = load_digits()
print([i for i in digits.target])

new_X = digits.data
train_y =digits.target
```

```python
# SCALING THE DATASET AND REDUCING DIMENSIONS TO 30 USING PCA
sc = StandardScaler()
new_X_train = sc.fit_transform(new_X)

pca = PCA(n_components = 30)
new_X_train = pca.fit_transform(new_X_train)

print(new_X_train.shape)
```

```python
# TRANSFORMING THE DATASET USING MULTI DIMENSIONAL SCALING
model = MDS(n_components=2)
X_mds = model.fit_transform(new_X_train)
print(X_mds.shape)
```

```python
# VISUALIZING THE DATASET RETURNED BY MULTIDIMENSIONAL SCALING TECHNIQUE
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=train_y, s=8)
plt.title('MDS Visualization')
plt.rcParams['figure.figsize'] = [10, 10]

plt.show()
```
